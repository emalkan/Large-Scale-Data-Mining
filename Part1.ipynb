```python
#Q1
import pandas as pd
import numpy as np

ratings = 'ml-latest-small/ratings.csv'
movies = 'ml-latest-small/movies.csv'

df_ratings = pd.read_csv(ratings)
given_ratings = len(df_ratings)
print('Possible ratings = ', given_ratings)

df_movies = pd.read_csv(movies)
num_movies = len(df_movies)
print('Total movies = ', num_movies)

num_users = len(np.unique(df_ratings.userId))
print('Users = ', num_users)

sparsity = float(given_ratings)/(num_movies * num_users)
print('Sparsity = ', sparsity)
```

```python
#Q2
import matplotlib.pyplot as plt

val, count = np.unique(df_ratings.rating, return_counts=True)
X = np.arange(0.5, 6, 0.5)
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Histogram of number of ratings per value")
plt.bar(val,count, width = 0.3)
plt.xticks(val, X)
plt.show()
```

```python
#Q3
val, count = np.unique(df_ratings.movieId, return_counts=True)
count1 = sorted(count, reverse = True)
plt.xlabel("Movie Index")
plt.ylabel("Number of ratings received")
plt.title("Histogram of number of ratings per movie")
plt.plot(count1)
plt.show()
```

```python
#Q4
val, count = np.unique(df_ratings.userId, return_counts=True)
count1 = sorted(count, reverse = True)
plt.xlabel("User Index")
plt.ylabel("Number of ratings received")
plt.title("Histogram of number of ratings per user")
plt.plot(count1)
plt.show()
```

```python
obj = df_ratings.groupby(by = 'movieId')
movies = obj.groups.keys()
V = obj['rating'].var(ddof=False)
```

```python
import matplotlib.pyplot as plt

range = np.arange(0, 6.1, 0.5)
temp = pd.cut(V, range, right=False, labels = range[0:len(range)-1])

val, count = np.unique(temp, return_counts=True)
plt.xlabel("Variance of rating values")
plt.ylabel("Number of movies")
plt.title("Number of movies with specified variance")
plt.bar(val,count, width = 0.3)
plt.xticks(val, range)
plt.show()
```
